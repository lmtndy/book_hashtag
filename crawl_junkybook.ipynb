```python
# Function to crawl book details
def crawl_book_details(url):
    book_data = []

    # Send a GET request to the web page
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Use BeautifulSoup to parse the HTML
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all elements containing book information
        book_wrappers = soup.find_all('div', class_='product-wrapper mb-40 hidden-sm')

        for book_wrapper in book_wrappers:
            # Find the child link within the product-wrapper element
            book_link = book_wrapper.find('a')['href']

            # Check if the link is relative or absolute
            if not book_link.startswith('http'):
                # If the link is relative, prepend the base URL
                book_link = urljoin(url, book_link)

            # Send a GET request to the book detail page
            book_response = requests.get(book_link)

            if book_response.status_code == 200:
                book_soup = BeautifulSoup(book_response.content, 'html.parser')

                # Extract book details from the detail page
                title = book_soup.find('div', class_='page-title').text.strip()
                description = book_soup.find('div', class_='product-social-links').text.strip()
                topic = book_soup.find('div', class_='reviews-actions').text.strip()

                book_data.append({'Title': title, 'Description': description, 'Topic': topic})
            else:
                print(f'Could not access book details.')
    else:
        print('Request was not successful.')

    return book_data
```

```python
# Crawl multiple pages and store book data in a list
all_books_data = []
for page_number in range(1, 38):
    page_url = f'https://www.junkybooks.com/books?page={page_number}'
    print(f'Crawling page {page_number}...')
    all_books_data.extend(crawl_book_details(page_url))

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(all_books_data)

# Write DataFrame to Excel file
excel_file_path = 'books_data.xlsx'
df.to_excel(excel_file_path, index=False)

print(f'Book data has been saved to {excel_file_path}.')
```
